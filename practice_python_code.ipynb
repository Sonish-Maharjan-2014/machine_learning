```python
random = np.zeros((1,m+1))
for i in range(1,m+1):
  seed = np.mod(a * seed + c,m)
  random[0,i] = seed/m
  # print(random[0,i])
# random.ndim,random.shape
# random[0,1]
```

```python
s = 'BANANA'
v = ['A','E','I','O','U']
```

```python
temp=[]
for i in range(len(s)):
  for j in range(len(s)+1):
    if i<j:
      temp.append(s[i:j])
```

```python
temp
```

```python
st = set(temp)
```

```python
d = {}
for i in st:
  d[i] = temp.count(i)
```

```python
s_score = 0
k_score = 0
for i in d:
  if i[0] not in v:
    s_score += d[i]
  else:
    k_score += d[i]
if s_score>k_score:
  print('Stuart',s_score)
else:
  print('Kevin',k_score)
```
