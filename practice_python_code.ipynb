```python
m   = 100
a = 6.3
c = 1.6
seed = 1
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random = np.zeros((1,m+1))
for i in range(1,m+1):
  seed = np.mod(a * seed + c,m)
  random[0,i] = seed/m
  # print(random[0,i])
# random.ndim,random.shape
# random[0,1]
```

```python
plt.plot(range(1,m+1),random[0,1:])
```

```python
n = 10
o1 = np.zeros((1,n+1))
ei = m/n
x2 = 0
```

```python
for i in range(1,m+1):
  temp = np.ceil(random[0,i]*10)
  o1[0,int(temp)] = o1[0,int(temp)]+1
# o1[0,1]
```

```python
o1
```

```python
for i in range(1,n+1):
  x2 = x2+((o1[0,i]-ei)**2/ei)
x2
```

```python
if __name__ == '__main__':
    N = int(input())
    temp = []
    for i in range(N):
        inp = list(str(input().split()))
        if inp[0] == 'insert':
            temp.insert(inp[1],inp[2])
        if inp[0] == 'remove':
            temp.remove(inp[1])
        if inp[0] == 'append':
            temp.append(inp[1])
        if inp[0] == 'sort':
            temp.sort()
        if inp[0] == 'pop':
            temp.pop()
        if inp[0] == 'reverse':
            temp.reverse()
        if inp[0] == 'print':
            print(temp)
```
