# Visualzation for Rank 3

```python
f = np.random.randn(3,3)
```

```python
np.linalg.matrix_rank(f)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
for i in range(100):
    
    in_ = np.random.randn(3)
    in_ = in_.T
    out = np.matmul(f,in_)
    out = out.T
    ax.scatter(in_[0],in_[1],in_[2],c='r',label = 'input')
    ax.scatter(out[0],out[1],out[2],c='b',label = 'output')
# ax.legend()
plt.show()
```

```python
np.linalg.inv(f)
```

```python
np.linalg.eig(f)
```
