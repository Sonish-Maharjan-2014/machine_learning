# predict y
### y = w1*x1 + w2*x2 + w3*x3 + b
### y = w dot x + b
### w =[w1, w2, w3]
### x = [x1, x2, x3]

```python
def cal_y_pred(w,b,x):
    return np.dot(w,x)+b
```

```python
cal_y_pred(w,b,x)
```

# Cost function

```python
def cal_cost_function(w,b,x):
    cost_function = (((cal_y_pred(w,b,x)-y)**2).sum())*(.5/m)
    return cost_function
```

# Gradient descent

```python
w = np.zeros(len(x))
b = np.zeros(1)
a = .0001016


xplot=[]
yplot=[]

while True:
    temp_w = []
    for j in range(len(x)):
        gradient_w = (((cal_y_pred(w,b,x)-y)*x[j]).sum())/m
        temp_w.append(w[j] - a*gradient_w)

    gradient_b = (((cal_y_pred(w,b,x)-y)).sum())/m
    temp_b = b - a*gradient_b

    print(gradient_b)
    if abs(gradient_b) < 0.001:
        break


    w = temp_w
    b = temp_b
```
