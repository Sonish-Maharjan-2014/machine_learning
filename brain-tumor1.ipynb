```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
```

```python
file_name = os.listdir(f'/kaggle/input/brain-tumor/Brain Tumor/Brain Tumor')
file_name.sort()
```

```python
path = '/kaggle/input/brain-tumor/Brain Tumor/Brain Tumor'
len(file_name)
```

```python
img = cv2.imread(f'{path}/{file_name[3]}',0)
img = cv2.resize(img,(128,128))
plt.imshow(img)
```

```python
image_array = []
for i in file_name:
    img = cv2.imread(f'{path}/{i}',0)
    img = cv2.resize(img,(128,128))
    image_array.append(img)
```

```python
plt.imshow(image_array[1])
```

```python
image_array = np.array(image_array)/255
```

```python
image_array = image_array.reshape(-1,128*128)
image_array.shape
```

```python
data_csv = pd.read_csv(f'/kaggle/input/brain-tumor/Brain Tumor.csv')
data_csv.head()
```

```python
d = data_csv.copy()
```

```python
d.columns
```

```python
col = ['Mean', 'Variance', 'Standard Deviation', 'Entropy',
       'Skewness', 'Kurtosis', 'Contrast', 'Energy', 'ASM', 'Homogeneity',
       'Dissimilarity', 'Correlation', 'Coarseness']
x = d[col]
```

```python
x1 = pd.DataFrame(image_array)
x1
```

```python
X = pd.concat([x1,x], axis =1)
```

```python
col_dict = {}
for i in range(16384):
    col_dict[i] = str(i)
```

```python
X.rename(columns=col_dict, inplace= True)
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from tensorflow.keras.layers import Conv2D,Input,Flatten,Dense,Softmax
from tensorflow.keras import Sequential
```

```python
X.shape
```

```python

# layer0 = Input(16397)
layer1 = Conv2D(filters = 32, kernel_size = 3,activation = 'relu', kernel_initializer = 'he_normal')
layer2 = Conv2D(filters = 64, kernel_size = 3,activation = 'relu', kernel_initializer = 'he_normal')
layer3 = Conv2D(filters = 128, kernel_size = 3,activation = 'relu', kernel_initializer = 'he_normal')
layer4 = Conv2D(filters = 256, kernel_size = 3,activation = 'relu', kernel_initializer = 'he_normal')
layer5 = Flatten()
layer6 = Dense(units = 50*50, activation = 'relu')
layer7 = Dense(units = 25*25, activation = 'relu')
layer8 = Dense(units = 10*10, activation = 'relu')
layer9 = Dense(units = 5*5, activation = 'relu')
layer10 = Dense(units = 2*2, activation = 'relu')
layer11 = Dense(units = 2, activation = 'softmax')
    
    
    
    
```

```python
k = [f'layer{i}' for i in range(12)]
```

```python
model = Sequential([
 layer1,
 layer2,
 layer3,
 layer4,
 layer5,
 layer6,
 layer7,
 layer8,
 layer9,
 layer10,
 layer11])
```

```python
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.optimizers import Adam
model.compile(optimizer = Adam(learning_rate = 0.001),
              loss = BinaryCrossentropy(),
              metrics = ['accuracy'])
```
